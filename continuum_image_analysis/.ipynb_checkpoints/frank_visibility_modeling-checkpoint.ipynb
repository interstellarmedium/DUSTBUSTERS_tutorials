```python
import os, sys, time
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
from astropy.visualization import (AsinhStretch, LogStretch, ImageNormalize)
from matplotlib.gridspec import GridSpec

%matplotlib inline            
```

```python
# set up plot parameters
import matplotlib as mpl
mpl.rc('xtick.major', size=5, pad=3, width=2)
mpl.rc('ytick.major', size=5, pad=3, width=2)
mpl.rc('xtick.minor', size=2, pad=3, width=2)
mpl.rc('ytick.minor', size=2, pad=3, width=2)
mpl.rc('axes', linewidth=2, labelsize=14, titlesize=18)
mpl.rc('legend', fontsize=14)
mpl.rc('lines', markersize=5)
mpl.rc('font', size=12)
cmap = 'magma'
```

```python
# geometric properties (from Huang et al. 2018 DSHARP II)
offRA, offDEC = +0.0019, -0.0025
incl, PA = 34.97, 85.76
inclr, PAr = np.radians(incl), np.radians(PA)

# CLEAN map header values needed for conversion from intensity to brightness temperature
# and data needed for comparison plot at the end
data = fits.open('data/AS209_continuum.fits')
image = np.squeeze(data[0].data)
header = data[0].header
nx, ny = header['NAXIS1'], header['NAXIS2']
RAo = 3600 * header['CDELT1'] * (np.arange(nx) - (header['CRPIX1'] - 1)) - offRA
DECo = 3600 * header['CDELT2'] * (np.arange(ny) - (header['CRPIX2'] - 1)) - offDEC
freq = header['CRVAL3']
beam_maj, beam_min, beam_PA = 3600 * header['BMAJ'], 3600 * header['BMIN'], header['BPA']
beam_area = (np.pi * beam_maj * beam_min / (4 * np.log(2))) / (3600 * 180 / np.pi)**2
rout = 1.4
```

## read in the visibiity data (downloaded from DSHARP) and fit with frank
#### see https://discsim.github.io/frank/

```python
from frank.radial_fitters import FrankFitter
from frank.geometry import FixedGeometry
from frank.io import save_fit
from frank.utilities import sweep_profile, convolve_profile

# load the visibility data
dat = np.load('data/AS209_continuum.npz')
# look at what's here and how to access the data
print(dat.files)
u, v, vis, wgt = dat['u'], dat['v'], dat['V'], dat['weights']

# set the disk viewing geometry
geom = FixedGeometry(incl, PA, dRA=offRA, dDec=offDEC)

# set the frankenstein hyper-parameters
Rmax  = 2.0    # this is ~1.5x the maximum radius where we see emission (in arcseconds)
Ncoll = 300    # this is the # of collocation points used in the model 
alpha = 1.3    # this controls how much noisy data is included in the fit   
wsmth = 0.1    # this controls how much the fit smooths over oscillations

# configure the fitting code setup
FF = FrankFitter(Rmax=Rmax, N=Ncoll, geometry=geom, alpha=alpha, weights_smooth=wsmth)

# fit the visibilities
sol = FF.fit(u, v, vis, wgt)

# save the fit
save_fit(u, v, vis, wgt, sol, prefix='AS209_continuum')
```

```python
# plot the radial profile and compare with the CLEAN map
r_frank = sol.r
Inu_frank = sol.mean

clean_beam = {'bmaj':beam_maj, 'bmin':beam_min, 'beam_pa':beam_PA}
Inu_frank_convolved = convolve_profile(sol.r, sol.mean, geom.inc, geom.PA, clean_beam)

# convert to brightness temperatures (R-J limit)
kB_, c_ = 1.38064852e-16, 2.99792e10
Tb_frank = c_**2 * 1e-23 * Inu_frank / (2 * kB_ * freq**2)
Tb_frank_convolved = c_**2 * 1e-23 * Inu_frank_convolved / (2 * kB_ * freq**2)

# read in the CLEAN map profile
r, Tb_data, eTb_data = np.loadtxt('AS209_radial_profile.txt').T

fig, ax = plt.subplots(figsize=(8,5))
ax.fill_between(r, Tb_data-eTb_data, Tb_data+eTb_data, color='gray', alpha=0.5)
ax.plot(r, Tb_data, 'k', lw=2, label='CLEAN')
ax.plot(r_frank, Tb_frank, 'r', lw=2, label='Frank fit')
ax.plot(r_frank, Tb_frank_convolved, 'b', lw=2, label='Frank convolved')
ax.set_xlim([0, rout])
ax.set_ylim([0.03, 40])
ax.set_yscale('log')
ax.set_yticks([0.1, 1, 10])
ax.set_yticklabels(['0.1', '1', '10'])
ax.set_xlabel('$r \,\,$ [arcsec]')
ax.set_ylabel('$T_b \,\,$ [K]')
ax.legend()
```

```python
fig.savefig('AS209_profiles.pdf')
```

## Image comparison (demonstration purposes; best done by subtracting the visibilities and cleaning the residuals)

```python
# sweep the frank profile to a 2d image
# some fudging of limits to make sure the output image size is the same as the data

dr = 3600*header['CDELT2']/2
xmax=RAo.max()+dr
ymax=DECo.max()

Tb_frank_2D, xmax, ymax = sweep_profile(sol.r, Tb_frank_convolved, xmax=xmax, ymax=ymax, dr=dr, project=True, phase_shift=False, geom=sol.geometry)
```

```python
print(image.shape)
print(Tb_frank_2D.shape)
```

```python
# intensity limits, and stretch
vmin, vmax = 0, 70    # these are in K units
norm = ImageNormalize(vmin=vmin, vmax=vmax, stretch=AsinhStretch(0.03))
im_bounds = (RAo.max(), RAo.min(), DECo.min(), DECo.max())
dRA_lims, dDEC_lims = [rout, -rout], [-rout, rout]

# image plot
gs = GridSpec(1,3, hspace=0.2, top=0.97, left=0.07, right=0.97)
fig = plt.figure(figsize=(16,7))
ax0 = fig.add_subplot(gs[0])
ax1 = fig.add_subplot(gs[1])
ax2 = fig.add_subplot(gs[2])
axes = [ax0, ax1, ax2]

Tb_data_2D = c_**2 * 1e-23 * image / beam_area / (2 * kB_ * freq**2)

im = ax0.imshow(Tb_data_2D, origin='lower', cmap=cmap, extent=im_bounds, norm=norm, aspect='equal')
imfit = ax1.imshow(Tb_frank_2D, origin='lower', cmap=cmap, extent=im_bounds, norm=norm, aspect='equal')
imres = ax2.imshow(100*(Tb_data_2D-Tb_frank_2D), origin='lower', cmap=cmap, extent=im_bounds, norm=norm, aspect='equal')

ax0.set_xlim(dRA_lims)
ax0.set_ylim(dDEC_lims)
ax0.set_xlabel('RA offset [arcsec]')
ax0.set_ylabel('DEC offset [arcsec]')
ax0.text(0.05,0.9, 'Data', color='white', fontsize=16, fontweight='bold', transform=ax0.transAxes)

ax1.set_xlim(dRA_lims)
ax1.set_ylim(dDEC_lims)
ax1.text(0.05,0.9, 'Frank', color='white', fontsize=16, fontweight='bold', transform=ax1.transAxes)

ax2.set_xlim(dRA_lims)
ax2.set_ylim(dDEC_lims)
ax2.text(0.05,0.9, 'Residual (x 100)', color='white', fontsize=16, fontweight='bold', transform=ax2.transAxes)
```
